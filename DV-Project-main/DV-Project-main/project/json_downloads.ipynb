```python

# Merge the DataFrames on 'team', aligning data for teams present across all seasons
# and filling in missing values for teams not present in some seasons.
dfs = [df2016, df2017, df2018, df2019]
df_merged = pd.concat(dfs).groupby('team.name', as_index=False).sum()

df_merged
```

```python
# Calculate the weighted sum for normalized points and normalized goalsDiff
# Adjust these weights based on your specific weights for each year
weights = [1, 2.5, 2.5**2, 2.5**3]  # Weights for 2016, 2017, 2018, 2019 seasons respectively

# Multiply normalized points and normalized goalsDiff by the respective weights
for i, year in enumerate(['2016', '2017', '2018', '2019']):
    df_merged[f'weighted_points_{year}'] = df_merged[f'points_normalized_{year}'] * weights[i]
    df_merged[f'weighted_goalsDiff_{year}'] = df_merged[f'goalsDiff_normalized_{year}'] * weights[i]

# Sum the weighted normalized points and goalsDiff across all years to get a total for each team
df_merged['total_weighted_points'] = df_merged[[f'weighted_points_{year}' for year in ['2016', '2017', '2018', '2019']]].sum(axis=1)
df_merged['total_weighted_goalsDiff'] = df_merged[[f'weighted_goalsDiff_{year}' for year in ['2016', '2017', '2018', '2019']]].sum(axis=1)

# Normalize the total weighted points and goalsDiff to get the team rating
max_weighted_points = df_merged['total_weighted_points'].max()
max_weighted_goalsDiff = df_merged['total_weighted_goalsDiff'].max()

df_merged['team_rating_points'] = df_merged['total_weighted_points'] / max_weighted_points
df_merged['team_rating_goalsDiff'] = df_merged['total_weighted_goalsDiff'] / max_weighted_goalsDiff

# Now df_merged contains the team ratings based on normalized points and goals difference.
```
